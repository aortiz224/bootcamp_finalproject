### Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import statsmodels.api as sm
```

```python
X= df_ml2[['creator','likes','total_units','rights','age','type_of_nft_GIF','type_of_nft_PHOTO','type_of_nft_VIDEO','nsfw_False',
      'nsfw_True']]
y= df_ml2['price_log10']
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X= df_ml2.drop(['price','price_log10'], axis=1)
X.head()
```

```python
y= df_ml2['price_log10']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size=0.80)
```

```python
# Creating, fitting, and scaling the data
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating the decision tree and fittin model
model = DecisionTreeRegressor(criterion='mse')
model.fit(X_train_scaled, y_train)
```

```python
# make predictions with the test data
predictions= model.predict(X_test_scaled)
```

```python
df= pd.DataFrame({'Actual':y_test, "Predicted":predictions})
df.sample(15)
```

```python
model.score(X_test_scaled, y_test)
```

```python
path= '../Resources/'
df.to_csv(path+'ml_dataframe.csv')
```
