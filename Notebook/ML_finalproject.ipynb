## Initial Data cleaning

```python
import pandas as pd
import numpy as np
import sqlalchemy
from config import database, user, password
```

```python
engine= sqlalchemy.create_engine(f'postgresql://{user}:{password}@localhost/{database}')
con= engine.connect()

nft_table= pd.read_sql("SELECT * FROM final_nft", con=engine)
nft_table
```

## Machine Learning

```python
df_ml= nft_table.copy()
df_ml
```

```python
df_ml['age']= 2022- df_ml['year_create']
df_ml
```

```python
df_ml=df_ml.drop(['year_create'], axis=1)
df_ml
```

```python
df_ml['price_log10']= np.log10(nft_table['price'])
df_ml
```

```python
df_ml=pd.get_dummies(df_ml, columns=['type_of_nft','nsfw','creator'])
df_ml
```

```python
df_ml2= df_ml.drop(['title','name_of_work','art_series'], axis=1)
df_ml2
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X= df_ml2.drop(['price','price_log10'], axis=1)
X.head()
```

```python
y= df_ml2['price_log10']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Creating, fitting, and scaling the data
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating the decision tree and fittin model
model = DecisionTreeRegressor(criterion='mse')
model.fit(X_train_scaled, y_train)
```

```python
# make predictions with the test data
predictions= model.predict(X_test_scaled)
```

```python
df= pd.DataFrame({'Actual':y_test, "Predicted":predictions})
df
```
