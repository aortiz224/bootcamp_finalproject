```python
results= []
for x in df_eda.type:
    if x == "PHOTO":
        results.append(2)
    elif x == "GIF":
        results.append(1)
    else:
        results.append(0)
df_eda['type_category']=results
df_eda.head()
```

```python
df_eda= df_eda.drop(columns=['type','cid','path','symbol', 'name','title','art_series'])
df_eda
```

```python
df_eda_encoded= pd.get_dummies(df_eda, columns=['creator','nsfw','year'])
df_eda_encoded
```
